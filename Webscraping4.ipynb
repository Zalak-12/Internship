## Scrape the details of most viewed videos on YouTube from Wikipedia

```python
driver=webdriver.Chrome(r'C:\Users\Zalak\Downloads\chromedriver_win32\chromedriver.exe')
url='https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
driver.get(url)
```

```python
Data=[]
try:
    name=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td")
    for i in name:
        Data.append(i.text)
except NoSuchElementException:
    Data.append('N/A')
except StaleElementReferenceException:
    Data.append('N/A')
```

```python
df=pd.DataFrame()
df['Rank']=[Data[i] for i in range(len(Data)) if i % 6 == 0]
df['Video Name']=[Data[i+1][0:-4] for i in range(len(Data)) if i % 6 == 0]
df['Artist']=[Data[i+2] for i in range(len(Data)) if i % 6 == 0]
df['Views (billions)']=[Data[i+3] for i in range(len(Data)) if i % 6 == 0]
df['Upload Date']=[Data[i+4] for i in range(len(Data)) if i % 6 == 0]
df.style.hide_index()
```
