## 6: Scrape data for first 100 sneakers you find when you visit flipkart.com andsearch for “sneakers” in the search field.

```python
url='https://www.flipkart.com/'
driver.get(url)
```

```python
search_job=driver.find_element_by_xpath('/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input')
search_job.send_keys('Sneakers')
search_btn=driver.find_element_by_xpath('/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/button')
search_btn.click()
```

```python
page_tag=driver.find_elements_by_xpath('//nav[@class="yFHi8N"]//a')
page=[]
for i in page_tag:
    page.append(i.get_attribute('href'))
```

```python
count=0
brand=[]
desp=[]
price=[]
dis=[]
while len(dis)<101:
    driver.get(page[count])
    brand_tag=driver.find_elements_by_xpath("//div[@class='_2WkVRV']")
    for i in brand_tag:
        brand.append(i.text)
    desp_tag=driver.find_elements_by_xpath("//div[@class='_2B099V']")
    for i in desp_tag:
        desp.append(i.text.split('\n')[1])
    price_tag=driver.find_elements_by_xpath("//div[@class='_30jeq3']")
    for i in price_tag:
        price.append(i.text)
    disc_tag=driver.find_elements_by_xpath("//div[@class='_3Ay6Sb']")
    for i in disc_tag:
        dis.append(i.text)
    count+=1
```

```python
flipkart_df=pd.DataFrame({'Brand NAME':brand[:100],'Description':desp[:100],'Price':price[:100],'Discount':dis[:100]})
flipkart_df
```

## Q7: Go to the link - https://www.myntra.com/shoes. Set Price filter to “Rs. 7149 to Rs. 14099 ” , Color filter to “Black”, as shown inthe below image.

```python
url='https://www.myntra.com/shoes'
driver.get(url)
```

```python
search_btn=driver.find_elements_by_xpath("//li[@class ='colour-listItem']//label[@class ='common-customCheckbox']/div[@class = 'common-checkboxIndicator']")
search_btn[0].click()
```

```python
search_btn=driver.find_elements_by_xpath("//ul[@class ='price-list']//label[@class ='common-customCheckbox vertical-filters-label']/div[@class = 'common-checkboxIndicator']")
search_btn[1].click()
```

```python
brand=[]
desp=[]
price=[]
while True:
    brand_tag=driver.find_elements_by_xpath("//div[@class= 'product-productMetaInfo']//h3[@class='product-brand']")
    for i in brand_tag:
        brand.append(i.text)
    desp_tag=driver.find_elements_by_xpath("//div[@class= 'product-productMetaInfo']//h4[@class='product-product']")
    for i in desp_tag:
        desp.append(i.text)
    price_tag=driver.find_elements_by_xpath("//div[@class= 'product-productMetaInfo']//div[@class='product-price']")
    for i in price_tag:
        price.append('Rs.'+i.text.split('Rs.')[1])
    if len(price)<100:
        driver.get(driver.find_element_by_xpath("//a[@rel = 'next']").get_attribute('href'))
    else:
        break
```

```python
flipkart_df=pd.DataFrame({'Brand NAME':brand[:100],'Description':desp[:100],'Price':price[:100]})
flipkart_df
```

## Q10: Write a python program to scrape the salary data for Data Scientist designation.

```python
url='https://www.ambitionbox.com'
driver.get(url)
```

```python
search_btn=driver.find_element_by_xpath('/html/body/div[1]/nav/nav/a[4]')
search_btn.click()
```

```python
search_job=driver.find_element_by_id('jobProfileSearchbox')
search_job.send_keys('Data Scientist')
```

```python
search_btn=driver.find_element_by_xpath('/html/body/div/div/div/main/section[1]/div[2]/div[1]/span/div/div/div[1]')
search_btn.click()
```

```python
rating=[]
days=[]
titles=[]
minmax=[]
titles_tags=driver.find_elements_by_xpath('//div[@class="name"]')
for  i in titles_tags:
    titles.append(i.text.split("\n"))
rating_tags=driver.find_elements_by_xpath('//div[@class="salaries sbold-list-header"]')
for  i in rating_tags:
    rating.append(i.text.split("\n"))
price_tags=driver.find_elements_by_xpath('//div[@class="average-indicator-wrapper"]//p')
for  i in price_tags:
    days.append(i.text)
minmax1=driver.find_elements_by_xpath('//div[@class="salary-values"]')
for  i in minmax1:
    minmax.append(i.text.split("\n"))
```

```python
ambition_df=pd.DataFrame({'NAME of the company':[item[0] for item in titles],
                          'Total Salary record':[item[1] for item in titles],
                          'Experience':[item[2] for item in rating],
                          'Min':[item[0] for item in minmax],
                          'Average':days,
                          'Max':[item[1] for item in minmax]})
ambition_df
```
